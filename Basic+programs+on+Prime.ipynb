1. Counting consecutive zero's occuring between two prime numbers

      In a Natural number set ,if a number is prime then it is printed and if it is not prime then that number is replaced by 0.

```python
#Prime Number program to generate prime numbers

#If the count of factors is < 2 then it is considered as prime number (as prime number has factors - 1 and itself)
count_of_factors = 0
#Creating a list to store output into list , this can be more easily done using list comprehension but elaborating steps in this case
numbers_list_all = []
#Creating another list to store only prime number
numbers_list_prime = []
#Reading input from the user (till how many numbers the user needs prime number)
end_number = int(input("The number till where you need to print prime numbers, the number starts from 1 to --- :"))

for i in range(1,end_number):
    for j in range(1,i):
        if i%j == 0:
            count_of_factors += 1
    if count_of_factors < 2:
        numbers_list_all.append(i)
        numbers_list_prime.append(i)
    else:
        numbers_list_all.append(0)    
    count_of_factors = 0
print("\n Output is stored in two above defined lists . So execute them in order to see the output")
    
```

To find the maximum consecutive zeros between two prime numbers 

We are using the below code to it

```python
#Storing the difference between primes in a list to find the maximum in the list .
difference_between_primes = []
number_of_elements_in_list = (len(numbers_list_prime)-1)
for i in range(0,number_of_elements_in_list):
    difference_between_consecutive_primes = abs(numbers_list_prime[i] - numbers_list_prime[i+1])
    difference_between_primes.append(difference_between_consecutive_primes)
    if difference_between_consecutive_primes > 1:
#This condition is used because there will be zero in the list only if the difference between two consecutive primes is greater than 1
        print(f"The consecutive zero's between primes {numbers_list_prime[i]} and {numbers_list_prime[i+1]} is {difference_between_consecutive_primes}\n")

#The below lines of code is to find out the maximum number of zeros between the sequence
maximum = max(difference_between_primes)

for index in range(0,len(difference_between_primes)):
    if difference_between_primes[index] == maximum:
        max_difference_between_consecutive_primes = abs(numbers_list_prime[index]-numbers_list_prime[index+1])
        print(f" The MAXIMUM consecutive zero's in the LIST is between {numbers_list_prime[index]} and {numbers_list_prime[index+1]} and the number of consecutive zero's is {max_difference_between_consecutive_primes}")
```

2.To print '*' for addition of every two consecutive numbers  

i.e., numbers = [1,2,3,4,5....]
 
    1+2 = 3 ---three '*' to be printed
    
    2+3 = 5 ---five '*' to be printed
    
    3+4 = 7 ---seven '*' to be printed
    
    ------- etc...

```python
input_number = int(input("Enter the number till where you want "))

for i in range(1,input_number):
    sum_of_conecutive_number = i + (i+1)
    #print("value of sum is",sum)
    print("*" * sum_of_conecutive_number)
```
